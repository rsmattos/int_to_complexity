```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# FUNÇÃO QUE CALCULA O A FUNÇÃO,
# RETORNANDO SE ELA CONVERGE OU QUANTOS PASSOS PARA DIVERGIR

def func(x, y):
    iterations = 0
    max_iter = 100
    
    tol = 0.005
    diverge = 1000
    norm = tol + 1
    
    z = complex(0,0)
    
    while ( iterations < max_iter ):
        f = z*z + complex(x,y)
        
        norm = abs( abs(f) - abs(z) )
        iterations += 1
        
        z = f
        
        if norm < tol:
            iterations = 0
            break
        elif norm > diverge:
            break
        
    return iterations

def func2(x, y):
    print(type(x))
    return x + y*1j
```

```python
# MANDELBROT SET

# montar o grid com diferentes valores para número real e imag
x_min = -2.5
x_max = 1
y_min = -1.2
y_max = 1.2

x = np.linspace(x_min, x_max, int(x_max-x_min)*1000)
y = np.linspace(y_min, y_max, int(x_max-x_min)*1000)

result = [[0 for i in range(len(x))] for j in range(len(y))]

# mandar para a função
for i in range(len(x)):
    for j in range(len(y)):
        result[j][i] = func(x[i],y[j])

#print(result)
```

```python
# plotting
size = 4
plt.figure(figsize=(19.2*size, 10.8*size), dpi=100)
plt.imshow(result, cmap=plt.cm.gnuplot2, aspect='auto')
plt.axis('off')

plt.savefig("mandelbrot.jpg", bbox_inches='tight',pad_inches=0)
```
